Problem 1: Warm Up – One-Dimension Random Walk along a line in a 2-D plane
Imagine a person walking down a road from point ‘A’ to point ‘B’. Ideally one expects that person would walk in a straight line. However, the person walks such that every step she moves forward, she also moves one step to the right or left with equal probability. Write a Python code to simulate this walk and plot it as follows:

a.	The user should input the number of steps, and the code should plot the position of the walker v.s. the step number.

b.	Run a loop that does the random walk simulation 100 times and for 20 steps for each simulation. Store the position of the walker in a list or an array for each step of each simulation. Plot the average position of the walker for each of the 20 steps taken, along with error bars. As the number of steps taken increases where is the walker most likely to be found at?

c.	Redo part(b) for the case with bias, i.e., where the probability of moving one step to the left slightly more – e.g., 5% more – than the probability of moving one step to the right.

Note: Put the appropriate label names and legend on the plot. Format and embellish the plot to make it easily understandable.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
steps = int(input("No of steps: "))
```

```python
def walky(steps):
  l = [-1,1]  # List of possible step directions: -1 (left) and 1 (right)
  pos = [0,0] # Initialize the starting position at (0, 0)
  positions=[]
  posx = []
  posy = []
  for i in range (0,steps):
    pos[0] = pos[0] + 1 # Move one step to the right

    x = np.random.choice(l,p=[0.5,0.5]) # Randomly choose left or right step with equal probability
    if x == 1:
      pos[1]=pos[1]+ 1 # move 1 step in y direction up
      positions.append(pos.copy())
      posx.append(pos[0])
      posy.append(pos[1])
    else:
      pos[1]=pos[1] - 1 # move 1 step in y direction down
      positions.append(pos.copy())
      posx.append(pos[0])
      posy.append(pos[1])


  return positions,posx,posy

pospos = walky(steps)[0]

xcord=[]
ycord = []
for i in  pospos:
  xcord.append(i[0]) # Extract horizontal coordinates
  ycord.append(i[1]) # Extract vertical coordinates

plt.scatter(xcord,ycord)
plt.grid()
plt.xlabel("Horizontal steps")
plt.ylabel("Vertical steps")
plt.title("Random Walking Problem")

```

Plot the average position of the walker for each of the 20 steps taken, along with error bars. As the number of steps taken increases where is the walker most likely to be found at?

```python
# Simulate 100 random walks of 20 steps each and collect the vertical positions
lit=[]
lit1=[]
for i in range(100):
  lit.append(walky(20)[1])
  lit1.append(walky(20)[2])
print(lit1)
# Calculate the mean vertical position at each step

mean = []
for i in range(len(lit1[0])):
  temp = 0
  for j in range (len(lit1)):
    temp = temp + lit1[j][i]

  val = temp/len(lit1)
  mean.append(val) #append value to mean
print(mean)

print(xcord)
plt.plot(xcord,mean)

```

```python
plt.errorbar(xcord,mean,yerr = np.std(mean)/np.sqrt(len(mean)))
plt.plot(xcord,mean)
```

```python
def poswalky(steps):
  l = [-1,1]
  pos = [0,0]
  positions=[]
  posx = []
  posy = []
  for i in range (0,steps):
    pos[0] = pos[0] + 1
# Randomly choose left or right step with different probabilities (0.55 for right, 0.45 for left)
    x = np.random.choice(l,p=[0.55,0.45])
  # Update the y-coordinate based on the chosen step direction
    if x == 1:
      pos[1]=pos[1]+ 1
      positions.append(pos.copy())
      posx.append(pos[0])
      posy.append(pos[1])
    else:
      pos[1]=pos[1] - 1
      positions.append(pos.copy())
      posx.append(pos[0])
      posy.append(pos[1])


  return positions,posx,posy

pospos = poswalky(steps)[0]

xcord=[]
ycord = []
for i in  pospos:
  xcord.append(i[0])
  ycord.append(i[1])

plt.scatter(xcord,ycord)
plt.grid()
plt.xlabel("Horizontal steps")
plt.ylabel("Vertical steps")
plt.title("Random Walking Problem")
```

```python
# Initialize empty lists to store x-coordinates and y-coordinates for 100 random walks
lit=[]
lit1=[]
# Perform 100 random walks with 20 steps each and collect x-coordinates and y-coordinates
for i in range(100):
  lit.append(poswalky(20)[1])
  lit1.append(poswalky(20)[2])
print(lit1)

# Calculate the mean (average) y-coordinate at each step across all 100 random walks
mean = []
for i in range(len(lit1[0])):
  temp = 0
  for j in range (len(lit1)):
    temp = temp + lit1[j][i]

  val = temp/len(lit1)
  mean.append(val)
print(mean)

print(xcord)
plt.plot(xcord,mean) # Create a line plot to visualize the mean y-coordinates at each step

```
