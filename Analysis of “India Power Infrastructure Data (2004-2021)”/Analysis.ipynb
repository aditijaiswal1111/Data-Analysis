Stratified Sampling is the process of categorizing data based on specific criteria to ensure a representative sample. In the context of this dataset, it guarantees the inclusion of both smaller and larger states in India. The key advantage of stratified sampling is its ability to provide a more accurate representation of the entire population. However, it requires upfront definition of strata, which can be a time-consuming task. On the other hand, in Random Sampling, there is no guarantee of balanced representation, and the sample may not accurately reflect the entire population, potentially leading to skewed results.

Cluster Summary Statistics:

xiv)	Stratify the states into three clusters based on their population size: small, medium, and large. Define the thresholds for each cluster.

xv)	Calculate the mean, median, and standard deviation of "Power Requirement Net Crore Units" for each cluster.

xvi)	Calculate the mean, median, and standard deviation of "Availability Of Power Net Crore Units" for each cluster.

xvii)	Create box plots to visualize the distribution of power requirements and availability within each cluster.

```python

# Define population thresholds for clusters
small_threshold = 0.01  # million
medium_threshold = 0.5  # million

# Categorize states into clusters based on population
def categorize_population(population):
    if population < small_threshold:
        return "Small"
    elif population < medium_threshold:
        return "Medium"
    else:
        return "Large"

df['Cluster'] = df['Population'].apply(categorize_population)

# Calculate mean, median, and standard deviation for "Power Requirement Net Crore Units" and "Availability Of Power Net Crore Units" for each cluster
cluster_summary = df.groupby('Cluster').agg({
    'Power_Requirement_Net_Crore_Units': ['mean', 'median', 'std'],
    'Availability_Of_Power_Net_Crore_Units': ['mean', 'median', 'std']
}).reset_index()

print(cluster_summary)
```

```python
# Create a new DataFrame 'df_2021' by filtering the original DataFrame 'df' to include only rows where the 'Year' column is '2020-21'.
df_2021 = df[df['Year'] == '2020-21']

# Calculate the sum of 'Population' for the filtered DataFrame 'df_2021' and multiply it by 10,000,000 (to convert to a larger unit, such as total population).
pop2020_21 = df_2021['Population'].sum()*10000000
```

```python
print('Predicted Power Requirements')


pop21_22 = 1361343000
pop22_23 = 1373761000
pop23_24 = 1386180000
```

```python
# Calculate the population change ratios for three consecutive years

pop_change_1 = (pop21_22 - pop2020_21)/pop2020_21
pop_change_2 = (pop22_23 - pop21_22)/pop21_22
pop_change_3 = (pop23_24 - pop22_23)/pop22_23

# Create new columns in the 'df_2021' DataFrame to represent power requirements for future years, starting from 2021-22


df_2021['Power Requirements 2021-22'] = df['Power_Requirement_Net_Crore_Units']*pop_change_1 + df['Power_Requirement_Net_Crore_Units']

df_2021['Power Requirements 2022-23'] = df_2021['Power Requirements 2021-22']*pop_change_2 + df_2021["Power Requirements 2021-22"]

df_2021['Power Requirements 2023-24'] = df_2021['Power Requirements 2022-23']*pop_change_3 + df_2021['Power Requirements 2022-23']


df_2021.head()
print(df_2021)
```
